```python
# Converting Gender Column as a Categorical Variable
gender=pd.get_dummies(df['Gender'],drop_first=True)
gender
```

```python
df.drop('Student_ID',axis=1,inplace=True)
```

```python
df.head()
```

```python
X=df.drop("Addicted_Score",axis=1)
y=df["Addicted_Score"]

# columns for applying linear regression
```

```python
from sklearn.model_selection import train_test_split
#Library for splitting data into test and train
from sklearn.linear_model import LinearRegression
```

```python
# features with non numeric values which can not be directly used by model
categorical_features = ['Gender', 'Academic_Level', 'Country',
                        'Most_Used_Platform', 'Affects_Academic_Performance',
                        'Relationship_Status']


X_encoded = pd.get_dummies(X, columns=categorical_features)
# converted categorical variables into dummy variables

X_train,X_test,y_train,y_test=train_test_split(X_encoded,y,test_size=0.2,random_state=0)
model=LinearRegression()
```

```python
model.fit(X_train,y_train)
```

```python
# Making predicitions
y_pred = model.predict(X_test)
```

```python
# Evaluating the performance

from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score

mse = mean_squared_error(y_test, y_pred)
mae = mean_absolute_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print("Mean Squared Error : ", mse)
print("Mean Absolute Error : ", mae)
print("R Score : ", r2)
```

```python
import matplotlib.pyplot as plt

plt.scatter(y_test, y_pred, color='blue')
plt.xlabel("Actual Values")
plt.ylabel("Predicted Values")
plt.title("Actual vs Predicted Values")
plt.grid(True)
plt.show()
```
